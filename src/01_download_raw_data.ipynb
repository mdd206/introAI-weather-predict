```python
# Import thư viện
from pathlib import Path

import pandas as pd
import requests
```

```python
# Cấu hình API và file lưu dữ liệu
url = "https://archive-api.open-meteo.com/v1/archive"
project_root = Path.cwd().parent if Path.cwd().name == "src" else Path.cwd()
output_file = project_root / "data" / "raw" / "hanoi_weather_2019_2025.csv"
output_file.parent.mkdir(parents=True, exist_ok=True)

params = {
    "latitude": 21.0285,
    "longitude": 105.8542,
    "start_date": "2019-01-01",
    "end_date": "2025-12-31",
    "hourly": ",".join([
        "precipitation",
        "relative_humidity_2m",
        "surface_pressure",
        "temperature_2m",
        "wind_speed_10m",
        "wind_direction_10m",
        "cloud_cover",
    ]),
    "timezone": "Asia/Ho_Chi_Minh",
    "wind_speed_unit": "kmh",
    "precipitation_unit": "mm",
}
```

```python
# Tải dữ liệu từ Open-Meteo
response = requests.get(url, params=params)
response.raise_for_status()

data = response.json()
df = pd.DataFrame(data["hourly"])
df = df.rename(columns={"time": "datetime"})
df["datetime"] = pd.to_datetime(df["datetime"])

df.head()
```

```python
# Lưu file raw và kiểm tra nhanh
df.to_csv(output_file, index=False)

print("Rows, columns:", df.shape)
print("Start time:", df["datetime"].min())
print("End time:", df["datetime"].max())
print("Saved file:", output_file)

df.tail()
```
